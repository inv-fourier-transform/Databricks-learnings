#### Parameter widgets

Widgets let you pass parameters (like date, mode, or path) to notebooks when running jobs.

```python
# Example - Add a date widget (Here, parameter name is 'run_date' and default value is 2026-01-01)

dbutils.widgets.text("run_date", "2026-01-01")
```

```python
# Read the widget value - This allows the same notebook to run for different dates in a job.

run_date = dbutils.widgets.get("run_date")
print(run_date)
```
