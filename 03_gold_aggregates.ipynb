Gold notebook

```python
from pyspark.sql import functions as F

df = spark.table("silver_oct_events")

gold = df.groupBy("event_type") \
    .agg(F.count("*").alias("event_count"))

gold.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold_events_by_type")
```

1) Reads Silver
2) Applies business aggregation
3) Writes Gold KPI table
