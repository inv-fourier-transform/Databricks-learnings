```python
from pyspark.sql.functions import col

df = spark.table("bronze_oct_events")

df_clean = df.filter(
    col("event_time").isNotNull() &
    col("user_id").isNotNull()
)

df_clean = df_clean.dropDuplicates(["user_session", "event_time"])

df_clean.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("silver_oct_events")
```

1) Reads Bronze
2) Removes bad rows
3) Removes duplicates
4) Writes clean Silver data
